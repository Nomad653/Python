```python
def sigmoid(X):
  a = []
  for item in X:
    a.append(1/(1+math.exp(-item)))
  return a
```

```python
import matplotlib.pyplot as plt

X = np.arange(-10.,10.,0.2)
sig = sigmoid(X)
plt.figure(figsize=(20,10))
plt.plot(X,sig)
plt.show()
```

```python
def sigmoid_derivative(X):
  return X*(1.0-X)
```

```python
x = np.arange(-10.,10.,0.2)
sig = sigmoid_derivative(x)
plt.figure(figsize=(20,10))
plt.plot(x,sig)
plt.show()
```
